```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 加载数据
X_new = pd.read_csv('C:/Users/RH/A_regression/ESM_embedding_320/data/data_ABTS_320.csv', delimiter=',', header=None, index_col=0, skiprows=1)
y_new = np.loadtxt('C:/Users/RH/A_regression/ESM_embedding_320/data/y_data_ABTS.csv', delimiter=',')

# 数据预处理
scaler = StandardScaler()
X_new_scaled = scaler.fit_transform(X_new)

# 划分训练集和验证集
X_train, X_val, y_train, y_val = train_test_split(X_new_scaled, y_new, test_size=0.2, random_state=42)

# 将输入数据转换为LSTM所需的三维形状 (samples, time steps, features)
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_val = np.reshape(X_val, (X_val.shape[0], 1, X_val.shape[1]))

# 定义LSTM模型
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, activation='relu', return_sequences=True, input_shape=(1, X_train.shape[2])),
    tf.keras.layers.LSTM(32, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# 编译模型
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
model.fit(X_train, y_train, epochs=100, batch_size=8, validation_data=(X_val, y_val))

# 加载测试集数据
X_out = pd.read_csv('C:/Users/RH/A_regression/ESM_embedding_320/data/X_test_ABTS_320.csv', delimiter=',', header=None, index_col=0, skiprows=1)
X_out_scaled = scaler.transform(X_out)
X_out_reshaped = np.reshape(X_out_scaled, (X_out_scaled.shape[0], 1, X_out_scaled.shape[1]))

# 在测试集上进行预测
predictions = model.predict(X_out_reshaped)

# 输出预测结果
print(predictions)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 加载数据
X_new = pd.read_csv('C:/Users/RH/A_regression/ESM_embedding_320/data/data_ABTS_320.csv', delimiter=',', header=None, index_col=0, skiprows=1)
y_new = np.loadtxt('C:/Users/RH/A_regression/ESM_embedding_320/data/y_data_ABTS.csv', delimiter=',')

# 数据预处理
scaler_X = StandardScaler()
scaler_y = MinMaxScaler()
X_new_scaled = scaler_X.fit_transform(X_new)
y_new_scaled = scaler_y.fit_transform(y_new.reshape(-1, 1))

# 划分训练集和验证集
X_train, X_val, y_train, y_val = train_test_split(X_new_scaled, y_new_scaled, test_size=0.2, random_state=42)

# 将输入数据转换为LSTM所需的三维形状 (samples, time steps, features)
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_val = np.reshape(X_val, (X_val.shape[0], 1, X_val.shape[1]))

# 定义LSTM模型
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, activation='relu', return_sequences=True, input_shape=(1, X_train.shape[2])),
    tf.keras.layers.LSTM(32, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# 编译模型
model.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val))

# 加载测试集数据
X_out = pd.read_csv('C:/Users/RH/A_regression/ESM_embedding_320/data/X_test_ABTS_320.csv', delimiter=',', header=None, index_col=0, skiprows=1)
X_out_scaled = scaler_X.transform(X_out)
X_out_reshaped = np.reshape(X_out_scaled, (X_out_scaled.shape[0], 1, X_out_scaled.shape[1]))

# 在测试集上进行预测
predictions_scaled = model.predict(X_out_reshaped)

# 反归一化预测结果
predictions = scaler_y.inverse_transform(predictions_scaled)

# 输出预测结果
print(predictions)
```
