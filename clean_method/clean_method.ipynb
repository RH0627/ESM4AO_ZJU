```python
from cleanlab.filter import find_label_issues
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import os
os.chdir('C:/Users/RH/A_try_change/ESM320_clean')
```

```python
df = pd.read_csv("C:/Users/RH/A_try_change/ESM320_clean/data/antioxidant_whole_esm2_t6_8M_UR50D_unified_320_dimension.csv")
row_to_delete = 1405
df = df[:row_to_delete+1]
df_c = df.copy()
df['label'] = preprocessing.LabelEncoder().fit_transform(df["label"])
df['label'] = df['label'].astype('category')

## Split data for evaluation
df.reset_index(drop=True, inplace=True)
```

```python
# Train model on noisy labels.
train_data = df.drop(['No', 'Sequence', 'label'], axis=1)
train_labels = df['label']
```

```python
# Get predicted probabilities through cross validation.
model = LogisticRegression(penalty='l2',max_iter=10000)
pred_probs = cross_val_predict(model, train_data, train_labels, method='predict_proba')

# Returns list of indices of label issues, sorted by self_confidence.
issue_idx = find_label_issues(train_labels, pred_probs, return_indices_ranked_by='self_confidence')
# Filter original data to show students with grade issues.
issue_stud_id = df.iloc[issue_idx].No.values.tolist()
issues_df = df_c[df_c['No'].isin(issue_stud_id)]
issues_df.sort_values(by="No", key=lambda column: column.map(lambda e: issue_stud_id.index(e)), inplace=True)
# Show a few good examples.
issues_df.head(-1)
```
